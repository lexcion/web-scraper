```python
import pandas as pd
import re

def is_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def clean_text(text):
    # Remove lines containing '+ '
    lines = text.strip().split('\n')
    cleaned_lines = [line for line in lines if not line.startswith('+ ')]
    return '\n'.join(cleaned_lines)

def parse_extracted_text(extracted_text):
    # Clean the text
    cleaned_text = clean_text(extracted_text)
    
    lines = cleaned_text.split('\n')
    data = []
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        
        if line == "Money Line":
            # Money Line parsing
            i += 1
            if i < len(lines):
                team1 = lines[i].strip()
                i += 1
            else:
                team1 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds1 = lines[i].strip()
                i += 1
            else:
                odds1 = None
            
            if i < len(lines):
                team2 = lines[i].strip()
                i += 1
            else:
                team2 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds2 = lines[i].strip()
                i += 1
            else:
                odds2 = None

            data.append(['Money Line', team1, None, odds1])
            data.append(['Money Line', team2, None, odds2])
        
        elif line == "Spread" or line == "Run Line":
            # Spread parsing
            i += 1
            if i < len(lines):
                team1 = lines[i].strip()
                i += 1
            else:
                team1 = None
            
            if i < len(lines) and (lines[i].strip().startswith('+') or lines[i].strip().startswith('-')):
                value1 = lines[i].strip()
                i += 1
            else:
                value1 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds1 = lines[i].strip()
                i += 1
            else:
                odds1 = None
            
            if i < len(lines):
                team2 = lines[i].strip()
                i += 1
            else:
                team2 = None
            
            if i < len(lines) and (lines[i].strip().startswith('+') or lines[i].strip().startswith('-')):
                value2 = lines[i].strip()
                i += 1
            else:
                value2 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds2 = lines[i].strip()
                i += 1
            else:
                odds2 = None

            data.append(['Spread', team1, value1, odds1])
            data.append(['Spread', team2, value2, odds2])
        
        elif line == "Total":
            # Total parsing
            i += 1
            if i < len(lines):
                outcome1 = lines[i].strip()
                i += 1
            else:
                outcome1 = None
            
            if i < len(lines) and is_float(lines[i].strip()) and float(lines[i].strip()) > 20:
                value1 = lines[i].strip()
                i += 1
            else:
                value1 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds1 = lines[i].strip()
                i += 1
            else:
                odds1 = None
            
            if i < len(lines):
                outcome2 = lines[i].strip()
                i += 1
            else:
                outcome2 = None
            
            if i < len(lines) and is_float(lines[i].strip()) and float(lines[i].strip()) > 20:
                value2 = lines[i].strip()
                i += 1
            else:
                value2 = None
            
            if i < len(lines) and is_float(lines[i].strip()):
                odds2 = lines[i].strip()
                i += 1
            else:
                odds2 = None

            data.append(['Total', outcome1, value1, odds1])
            data.append(['Total', outcome2, value2, odds2])
        
        else:
            i += 1
    
    # Create DataFrame
    df = pd.DataFrame(data, columns=['Bet Type', 'Team/Outcome', 'Value', 'Odds'])
    return df

df1 = parse_extracted_text(extracted_text3)
print("Parsed DataFrame from extracted_text3:")
print(df1)
```

```python
import pandas as pd
import re

def extract_game_state(text, keyword):
    # Normalize the text and split into lines
    lines = text.split("\n")
    
    # Find the line containing the keyword within context using regex
    keyword_pattern = re.compile(re.escape(keyword), re.IGNORECASE)
    for i, line in enumerate(lines):
        if keyword_pattern.search(line.strip()):
            # Return the remaining lines starting from the keyword line
            return lines[i:]
    return []

import pandas as pd
import re

def parse_moneyline(lines, game_state):
    data = []
    moneyline_detected = False
    
    for line in lines:
        if "Money Line" in line:
            moneyline_detected = True
            continue
        
        if moneyline_detected:
            team_name = line.strip()
            odds_line = lines[lines.index(line) + 1] if lines.index(line) + 1 < len(lines) else ""
            
            if re.match(r'\d+\.\d+', odds_line):
                odds = odds_line.strip()
                data.append({
                    'Game State': game_state,
                    'Bet Type': 'Money Line',
                    'Team/Outcome': team_name,
                    'Value': None,
                    'Odds': odds
                })
            else:
                continue  # Skip if odds are not found
            
            # Add second team data if available
            if lines.index(line) + 2 < len(lines):
                next_team_name = lines[lines.index(line) + 2].strip()
                next_odds_line = lines[lines.index(line) + 3] if lines.index(line) + 3 < len(lines) else ""
                
                if re.match(r'\d+\.\d+', next_odds_line):
                    next_odds = next_odds_line.strip()
                    data.append({
                        'Game State': game_state,
                        'Bet Type': 'Money Line',
                        'Team/Outcome': next_team_name,
                        'Value': None,
                        'Odds': next_odds
                    })
            
            moneyline_detected = False  # Reset flag
    
    return pd.DataFrame(data)

```

```python
import pandas as pd
import re

def parse_decimal_value_odds(value_string):
    match = re.search(r'(\d+\.\d+)\s*(\d+\.\d+)?', value_string)
    if match:
        value = match.group(1)
        odds = match.group(2) if match.group(2) else None
        return value, odds
    return None, None

def parse_total(lines, game_state):
    data = []
    total_detected = False
    
    for i, line in enumerate(lines):
        if "Total" in line:
            total_detected = True
            continue
        
        if total_detected:
            outcome = line.strip()
            next_line = lines[i + 1] if i + 1 < len(lines) else ""
            value, odds = parse_decimal_value_odds(next_line)
            
            if value and odds:
                data.append({
                    'Game State': game_state,
                    'Bet Type': 'Total',
                    'Team/Outcome': outcome,
                    'Value': value,
                    'Odds': odds
                })
                total_detected = False
                continue
            elif not value:
                next_line = lines[i + 2] if i + 2 < len(lines) else ""
                value, odds = parse_decimal_value_odds(next_line)
                if value and odds:
                    data.append({
                        'Game State': game_state,
                        'Bet Type': 'Total',
                        'Team/Outcome': outcome,
                        'Value': value,
                        'Odds': odds
                    })
                total_detected = False
    
    return pd.DataFrame(data)
```

```python
import pandas as pd
import re

def parse_spread(lines, game_state):
    data = []
    spread_detected = False
    
    for line in lines:
        if "Spread" in line or "Run Line" in line:
            spread_detected = True
            continue
        
        if spread_detected:
            team_name = line.strip()
            odds_line = lines[lines.index(line) + 1] if lines.index(line) + 1 < len(lines) else ""
            
            if re.match(r'\d+\.\d+', odds_line):
                odds = odds_line.strip()
                value = None
                data.append({
                    'Game State': game_state,
                    'Bet Type': 'Spread',
                    'Team/Outcome': team_name,
                    'Value': value,
                    'Odds': odds
                })
            else:
                continue
            
            # Add second team data if available
            if lines.index(line) + 2 < len(lines):
                next_team_name = lines[lines.index(line) + 2].strip()
                next_odds_line = lines[lines.index(line) + 3] if lines.index(line) + 3 < len(lines) else ""
                
                if re.match(r'\d+\.\d+', next_odds_line):
                    next_odds = next_odds_line.strip()
                    data.append({
                        'Game State': game_state,
                        'Bet Type': 'Spread',
                        'Team/Outcome': next_team_name,
                        'Value': value,
                        'Odds': next_odds
                    })
            
            spread_detected = False
    
    return pd.DataFrame(data)
```

```python
# Update getMainOdds to include parse_spread
def getBMMainOdds(extracted_text):
    lines = extract_game_state(extracted_text, keyword="Game")
    
    df_moneyline = parse_moneyline(lines, game_state="Main")
    df_total = parse_total(lines, game_state="Main")
    df_spread = parse_spread(lines, game_state="Main")
    
    df_combined = pd.concat([df_moneyline, df_total, df_spread], ignore_index=True)
    return df_combined

# Define getH1Odds to extract first half odds
def getBMH1Odds(extracted_text):
    lines = extract_game_state(extracted_text, keyword="First Half")
    
    df_moneyline = parse_moneyline(lines, game_state="First Half")
    df_total = parse_total(lines, game_state="First Half")
    df_spread = parse_spread(lines, game_state="First Half")
    
    df_combined = pd.concat([df_moneyline, df_total, df_spread], ignore_index=True)
    return df_combined

# Define getH2Odds to extract second half odds
def getBMH2Odds(extracted_text):
    lines = extract_game_state(extracted_text, keyword="Second Half")
    
    df_moneyline = parse_moneyline(lines, game_state="Second Half")
    df_total = parse_total(lines, game_state="Second Half")
    df_spread = parse_spread(lines, game_state="Second Half")
    
    df_combined = pd.concat([df_moneyline, df_total, df_spread], ignore_index=True)
    return df_combined
```
