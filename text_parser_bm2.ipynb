```python
import re

def clean_text(text):
    # Define a list of phrases and words to remove
    phrases_to_remove = ["Total","Spread","Run Line","Back to Schedule", "Money Line", "Bet Slip","Action", "@", "vs","My Open Bets","Empty","Select any pick to start betting","LIVE CHAT"]
    
    # Split the text into lines
    lines = text.splitlines()

    # Remove specified phrases and empty lines
    cleaned_lines = []
    i=0
    for line in lines:
        i=i+1
        # Remove specified phrases
        for phrase in phrases_to_remove:
            line = line.replace(phrase, "")
        
        # Remove extra spaces
        line = re.sub(r'\s+', ' ', line).strip()
        
        # Skip lines that are empty or start with "+ "
        if not line or line.startswith("+ "):
            continue
        
        if(i < len(lines)):
            if re.search(r'^\d+(\.\d+)?$', lines[i]):
                cleaned_lines.append(line)
                continue
            
        # Check if the line contains more than 4 digits
        if len(re.findall(r'\d', line)) == 3 and not line.startswith("+") and not line.startswith("-"):
            # Extract the last four characters as odds
            odds = line[-4:]
            # Extract the remaining part as value
            value = line[:-4].strip()
            # Add the cleaned line with value and odds
            cleaned_lines.append(f"{value} {odds}")
        
        elif len(re.findall(r'\d', line)) >= 4 and not line.startswith("+") and not line.startswith("-"):
            # Extract the last four characters as odds
                
            odds = line[-5:]
            try:
                oddfloat = 0
                oddfloat = float(odds)
            except:
                oddfloat = 0
            if (oddfloat > 20):
                odds = line[-4:]
                value = line[:-4].strip()
            # Extract the remaining part as value
            else:
                value = line[:-5].strip()
            # Add the cleaned line with value and odds
            cleaned_lines.append(f"{value} {odds}")
        elif len(re.findall(r'\d', line)) >=4:
            # Extract the last four characters as odds
            odds = line[-4:]
            # Extract the remaining part as value
            value = line[:-4].strip()
            # Add the cleaned line with value and odds
            cleaned_lines.append(f"{value} {odds}")
        else:
            cleaned_lines.append(line)

    # Combine consecutive lines that contain numbers, but only if they don't contain letters
    combined_lines = []
    i = 0
    while i < len(cleaned_lines):
        line = cleaned_lines[i].strip()
        
        # Check if the line contains only numbers and possibly + or - (i.e., no letters)
        if re.search(r'^\d+(\.\d+)?$', line) or re.search(r'^[+-]\d+(\.\d+)?$', line):
            # If the current line contains only numbers, merge it with the previous line
            if combined_lines and not re.search(r'[a-zA-Z]', combined_lines[-1]):
                combined_lines[-1] += f" {line}"
            else:
                combined_lines.append(line)
        else:
            combined_lines.append(line)
        
        i += 1
    if combined_lines[0] == "0":
        combined_lines = combined_lines[1:]
    # Join the cleaned lines back into a single string with newline characters
    cleaned_text = "\n".join(combined_lines)
    
    return cleaned_text
```

```python
import pandas as pd

def matchinfo(cleaned_text):
    # Split the cleaned text into lines
    lines = cleaned_text.splitlines()

    # Initialize variables
    team_1 = team_2 = status = date = "N/A"
    
    # Extract team names and game status/date
    if len(lines) >= 2:
        team_1 = lines[0].strip()
        team_2 = lines[1].strip()

    if len(lines) >= 3:
        third_line = lines[2].strip()
        if third_line == "Live Game":
            status = "live"
            date = ""
        elif third_line and third_line[0].isdigit():
            date = third_line.replace(" ", "")  # Remove any spaces in date
            if len(date) >= 5:
                date = date[:5] + " " + date[5:]  # Separate the first 5 characters from the rest
            status = "pregame"
        else:
            status = "N/A"
            date = "N/A"

    # Create a DataFrame with the extracted information
    df = pd.DataFrame({
        "team 1": [team_1],
        "team 2": [team_2],
        "status": [status],
        "date": [date]
    })

    return df
```

```python
import re
import pandas as pd

def formatodds(df, text):
    # Extract team names from the DataFrame
    team_1 = "!"+df['team 1'].iloc[0]
    team_2 = "!"+df['team 2'].iloc[0]
    
    start_index = 0
    
    # Split the text into lines
    lines = text.splitlines()
    
    
    # Keep lines from "!Game" onwards
    lines = lines[start_index:]
    
    formatted_lines = []
    i = 0
    
    def format_line(line):
        """Format a line based on the given rules."""
        if re.search(r'\d+\.\d+', line):
            if ' ' in line:
                # Contains a number with space
                line = f";;{line.replace(' ', ';;')}"
            else:
                # Contains a number with no space
                line = f";;0;;{line}"
        return line
    
    # First pass: Process each line
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        
        if re.search(r'\d', line) and re.search(r'\.', line) and not re.search('[a-zA-Z]', line):
            # Format the line
            formatted_line = format_line(line)
            
            # Concatenate with the previous line if it doesn't already have a comma
            if formatted_lines and not re.search(r';;', formatted_lines[-1]):
                formatted_lines[-1] += formatted_line
            else:
                formatted_lines.append(formatted_line)
        else:
            # Add lines without numbers or not in the desired format as is
            formatted_lines.append(line)
        
        i += 1
        
    # Second pass: Add '!' in front of lines that don't contain both ',' and '.'
    final_lines = []
    for line in formatted_lines:
        if not (';;' in line and '.' in line):
            final_lines.append(f"!{line}")
        else:
            final_lines.append(line)
    
     # Find the index of the line containing "!Game"
    
    for x, line in enumerate(final_lines):
        if line == "!Game":
            start_index = x
            break

    
    if start_index is None:
        return ""  # Return an empty string if "!Game" is not found
    
    # Remove the two lines under "!Game" if they match team_1 or team_2
    i=0
    
    cleaned_lines = []
    for line in final_lines[0:]:
        stripped_line = line.strip()
        if stripped_line not in [team_1, team_2, "!Over", "!Under"]:
            cleaned_lines.append(stripped_line)
    
    
    # Join the final lines back into a single string with newline characters
    formatted_text = "\n".join(cleaned_lines)
    
    return formatted_text
```

```python
def clean_event_names(df):
    def clean_name(name):
        if "First Half" in name or "First half" in name or "FIRST HALF" in name or "1st Half" in name or "1st half" in name:
            return "H1"
        elif "SECOND HALF" in name or"Second Half" in name or "Second half" in name or "2nd Half" in name or "2nd half" in name:
            return "H2"
        elif "1st Quarter" in name or "FIRST QUARTER" in name or "First Quarter" in name or "First quarter" in name:
            return "Q1"
        elif "2nd Quarter" in name or "SECOND QUARTER" in name or "Second Quarter" in name or "Second quarter" in name:
            return "Q2"
        elif "3rd Quarter" in name or "THIRD QUARTER" in name or "Third Quarter" in name or "Third quarter" in name:
            return "Q3"
        elif "4th Quarter" in name or "FOURTH QUARTER" in name or "Fourth Quarter" in name or "Fourth quarter" in name:
            return "Q3"
        else:
            return name
    
    df['Event Name'] = df['Event Name'].apply(clean_name)
    return df

def store_dataframe(formatted_text,df):
    # Split the formatted text into lines
    lines = formatted_text.splitlines()
    
    # Extract team names from the DataFrame
    team_1 = df['team 1'].iloc[0]
    team_2 = df['team 2'].iloc[0]
    
    # Initialize lists to store the data
    event_names = []
    outcomes = []
    values = []
    odds_list = []
    
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("!"):
            
            # Process the event name
            event_name = line[1:].strip()
            
            # Check if the next line also starts with "!"
            if i + 1 < len(lines) and lines[i + 1].strip().startswith("!"):
                i += 1  # Skip the next line
                continue
            
            # Prepare to read lines until the next "!" or end of the text
            i += 1
            while i < len(lines) and not lines[i].strip().startswith("!"):
                next_line = lines[i].strip()
                if ';;' in next_line and '.' in next_line:
                    # Split the line by ","
                    parts = next_line.split(';;')
                    if len(parts) >= 3:
                        outcome = parts[0].strip()
                        value = parts[1].strip()
                        odds_value = parts[2].strip()
                        
                        # Append to lists
                        event_names.append(event_name)
                        outcomes.append(outcome)
                        values.append(value)
                        odds_list.append(odds_value)
                i += 1
        else:
            i += 1
    
    # Create a DataFrame from the lists
    df = pd.DataFrame({
        'Event Name': event_names,
        'Outcome': outcomes,
        'Value': values,
        'Odds': odds_list
    })
    
    # Clean the event names
    df = clean_event_names(df)
    
    return df
```

```python
def parse_bm_text(text):
    text = clean_text(text)
    info = matchinfo(text)
    formatted = formatodds(info,text)
    df = store_dataframe(formatted,info)
    return df
```
